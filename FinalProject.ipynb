```python
from pybaseball import standings
import pandas as pd
from datascience import *

from pybaseball import team_pitching
from pybaseball import team_batting

from pybaseball import pitching_stats
from pybaseball import batting_stats
```

```python
class TeamStandings:
# The TeamStandings class allows the user to see the statistics on the Team Standings.
  def __init__(self):
    self.data = standings(2024)
    self.data = pd.concat(self.data)
    self.standings_table = Table.from_df(self.data)
    standings_table_df = self.standings_table.to_df()
    standings_table_df = standings_table_df.replace('--', '0.0', regex=True)
    standings_table_df['GB'] = pd.to_numeric(standings_table_df['GB'], errors='coerce')
    self.standings_table = Table.from_df(standings_table_df)

# The showTable function allows the user to see the table.
  def showTable(self):
    self.standings_table.show()

# The sortTable function sorts the table for the stat that the user wants.
  def sortTable(self, stat):
    sorted_Table = self.standings_table.sort(stat, descending=True)
    sorted_Table.show()

# The filterTable function filters the table with what the user wants and then sorts the stats that is being filtered.
  def filterTable(self, stat, low, high):
    filtered_table = self.standings_table.copy()
    filtered_table = filtered_table.where(0, are.equal_to(0))
    for row in self.standings_table.rows:
      num = row.item(stat)
      if low <= float(num) <= high:
        filtered_table.append(row)
    filtered_table = filtered_table.sort(stat, descending=True)
    filtered_table.show()

class TeamPitchingStats:
# The TeamPitchingStats class allows the user to see the statistics for each team's pitching.
  def __init__(self):
    self.data = team_pitching(2024)
    selected_columns = ['Team', 'W', 'L', 'ERA', 'CG', 'SV', 'BS', 'IP', 'H', 'BB', 'SO', 'WHIP']
    self.data = self.data[selected_columns]
    self.pitching_stats_table = Table.from_df(self.data)

# The showTable function allows the user to see the table.
  def showTable(self):
    self.pitching_stats_table.show()

# The sortTable function sorts the table for the stat that the user wants.
  def sortTable(self, stat):
    sorted_Table = self.pitching_stats_table.sort(stat, descending=True)
    sorted_Table.show()

# The filterTable function filters the table with what the user wants and then sorts the stats that is being filtered.
  def filterTable(self, stat, low, high):
    filtered_table = self.pitching_stats_table.copy()
    filtered_table = filtered_table.where(0, are.equal_to(0))
    for row in self.pitching_stats_table.rows:
      num = row.item(stat)
      if low <= float(num) <= high:
        filtered_table.append(row)
    filtered_table = filtered_table.sort(stat, descending=True)
    filtered_table.show()

class IndividualPitchingStats(TeamPitchingStats):
# The IndividualPitchingsStats class allows the user to see the statistics for each qualified player's pitching.
  def __init__(self):
    self.data = pitching_stats(2024)
    selected_columns = ['Team','Name', 'W', 'L', 'ERA', 'CG', 'IP', 'H', 'BB', 'SO', 'WHIP']
    self.data = self.data[selected_columns]
    self.pitching_stats_table = Table.from_df(self.data)

class TeamBattingStats:
# The TeamBattingStats class allows the user to see the statitsics for each team's batting.
  def __init__(self):
    self.data = team_batting(2024)
    selected_columns = ['Team', 'AB', 'PA', 'H', '2B', 'HR', 'R', 'RBI', 'BB', 'SO', 'HBP', 'SB', 'CS']
    self.data = self.data[selected_columns]
    self.batting_stats_table = Table.from_df(self.data)

# The showTable function allows the user to see the table.
  def showTable(self):
    self.batting_stats_table.show()

# The sortTable function sorts the table for the stat that the user wants.
  def sortTable(self, stat):
    sorted_Table = self.batting_stats_table.sort(stat, descending=True)
    sorted_Table.show()

# The filterTable function filters the table with what the user wants and then sorts the stats that is being filtered.
  def filterTable(self, stat, low, high):
    filtered_table = self.batting_stats_table.copy()
    filtered_table = filtered_table.where(0, are.equal_to(0))
    for row in self.batting_stats_table.rows:
      num = row.item(stat)
      if low <= float(num) <= high:
        filtered_table.append(row)
    filtered_table = filtered_table.sort(stat, descending=True)
    filtered_table.show()

class IndividualBattingStats(TeamBattingStats):
# The IndividualBattingStats class allows the user to see the statistics for each qualified player's batting.
  def __init__(self):
    self.data = batting_stats(2024)
    selected_columns = ['Team','Name', 'AB', 'PA', 'H', '2B', 'HR', 'R', 'RBI', 'BB', 'SO', 'HBP', 'SB', 'CS']
    self.data = self.data[selected_columns]
    self.batting_stats_table = Table.from_df(self.data)
```

```python
Team_Standings_Table = TeamStandings()
Team_Pitching_Table = TeamPitchingStats()
Team_Batting_Table = TeamBattingStats()
Individual_Pitching_Table = IndividualPitchingStats()
Individual_Batting_Table = IndividualBattingStats()

# The process_standings function shows the user the standings table depending on how they want to sort and filter it.
def process_standings():
  Team_Standings_Table.showTable()
  input3 = input("Which stat do you want to sort by? (W/L/W-L%/GB/stop): ").upper()
  if input3.lower() == "stop":
    return
  Team_Standings_Table.sortTable(input3)
  input4 = input("In between what do you want to filter by? (lowerbound-higherbound/stop): ").lower()
  if input4.lower() == "stop":
    return
  lower_bound = float(input4.split("-")[0])
  higher_bound = float(input4.split("-")[1])
  Team_Standings_Table.filterTable(input3, lower_bound, higher_bound)

# The process_team_pitching function shows the user the team pitching table depending on how they want to sort and filter it.
def process_team_pitching():
  Team_Pitching_Table.showTable()
  input3 = input("Which stat do you want to sort by? (W/L/ERA/CG/SV/BS/IP/H/BB/SO/WHIP): ").upper()
  if input3.lower() == "stop":
    return
  Team_Pitching_Table.sortTable(input3)
  input4 = input("In between what do you want to filter by? (lowerbound-higherbound): ")
  if input4.lower() == "stop":
    return
  lower_bound = float(input4.split("-")[0])
  higher_bound = float(input4.split("-")[1])
  Team_Pitching_Table.filterTable(input3, lower_bound, higher_bound)

# The process_team_batting function shows the user the team batting table depending on how they want to sort and filter it.
def process_team_batting():
  Team_Batting_Table.showTable()
  input3 = input("Which stat do you want to sort by? (Team/AB/PA/H/2B/HR/R/RBI/BB/SO/HBP/SB/CS): ").upper()
  if input3.lower() == "stop":
    return
  Team_Batting_Table.sortTable(input3)
  input4 = input("In between what do you want to filter by? (lowerbound-higherbound): ")
  if input4.lower() == "stop":
    return
  lower_bound = float(input4.split("-")[0])
  higher_bound = float(input4.split("-")[1])
  Team_Standings_Table.filterTable(input3, lower_bound, higher_bound)

# The process_individual_pitching function shows the user the individual pitching table depending on how they want to sort and filter it.
def process_individual_pitching():
  Individual_Pitching_Table.showTable()
  input3 = input("Which stat do you want to sort by? (W/L/ERA/CG/IP/H/BB/SO/WHIP): ").upper()
  if input3.lower() == "stop":
    return
  Individual_Pitching_Table.sortTable(input3)
  input4 = input("In between what do you want to filter by? (lowerbound-higherbound): ")
  if input4.lower() == "stop":
    return
  lower_bound = float(input4.split("-")[0])
  higher_bound = float(input4.split("-")[1])
  Individual_Pitching_Table.filterTable(input3, lower_bound, higher_bound)

# The process_individual_batting function shows the user the individual batting table depending on how they want to sort and filter it.
def process_individual_batting():
  Individual_Batting_Table.showTable()
  input3 = input("Which stat do you want to sort by? (Team/AB/PA/H/2B/HR/R/RBI/BB/SO/HBP/SB/CS): ").upper()
  if input3.lower() == "stop":
    return
  Individual_Batting_Table.sortTable(input3)
  input4 = input("In between what do you want to filter by? (lowerbound-higherbound): ")
  if input4.lower() == "stop":
    return
  lower_bound = float(input4.split("-")[0])
  higher_bound = float(input4.split("-")[1])
  Individual_Standings_Table.filterTable(input3, lower_bound, higher_bound)

```
